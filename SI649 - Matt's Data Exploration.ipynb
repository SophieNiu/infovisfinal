The climate opinion data has 435 IDs, which is the expected number. I'll use that as a guide for cleaning up the other files

```python
#Master list of Ids from Climate Opinion data, formatted as text
DistIds = sorted(['{:04}'.format(i) for i in clopinion['GEOID'].unique()])
```

```python
#Which of the Climate Opinion Ids are missing from Dr. Samson's data?
[x for x in DistIds if x not in ['{:04}'.format(i) for i in devents['districtId'].unique()]]
```

Each of these districts seem to legitimately be missing from Dr. Samson's data, in spite of most of them having been around for decades... I think we can consider these to be districts with missing data.


```python
#Which of Dr. Samson's Ids are missing from the Climate Opinion data?
['{:04}'.format(i) for i in devents['districtId'].unique() if '{:04}'.format(i) not in DistIds]
```

```python
#Which are missing from the Congressional GeoJSON data?
[x for x in DistIds if x not in congprop['geoid'].unique().tolist()]
```

```python
#Which districts in the Congressional GeoJSON data are missing from the Climate Opinion Data?
BadGeoDist = [x for x in congprop['geoid'].unique().tolist() if x not in DistIds]
BadGeoDist
```

```python
#Let's Drop those!
congprop = congprop[~congprop['geoid'].isin(BadGeoDist)]
```

```python
#Looks like we are still 1 record over 435...
congprop.shape
```

```python
#Searching for districts in the json file with duplicate representation data
congprop.groupby('geoid',as_index=False)['geoid'].agg({'count':'count'}).query('count > 1').sort_values('count',ascending=False)
```

```python
congprop[congprop['geoid'] == '2801']
```

[Wikipedia](https://en.wikipedia.org/wiki/Mississippi%27s_1st_congressional_district) tells me that Alan Nunnelee passed away, and Trent Kelly took over the seat. I'll drop Alan Nunnelee and retain Trent Kelly.

```python
congprop = congprop[congprop['name']!='Nunnelee, Alan']
```

```python
congprop.shape
```

### Putting together cleaned-up Climate Opinion & GeoJSON Datasets

__Climate Data Merged w/ Congressional Info__

```python
#Confirming that the Ids match between the two datasets after my transformations
[x for x in congprop['geoid'].unique() if x not in DistIds]
```

```python
congred = congprop[['geoid','rep_sen','name','party','state_code','district','state_label','url']]
congred['geoid'] = congred['geoid'].astype('int64')
congred.head()
```

```python
clopinion.shape
```

```python
cong_opinion = clopinion.merge(congred,left_on='GEOID',right_on='geoid',how='left')
```

```python
cong_opinion.to_csv('dataset/CleanedData/Congress_YCOM_2019_Data.csv',index=False)
```

__GeoJSON File - Integer GeoIds, extra districts removed__

```python
#Filtering out the unneccessary records discovered above
geoupdate = [x for x in geofile['features'] if x['properties']['rep_sen'] == 'Representative' 
             and x['properties']['name'] != 'Nunnelee, Alan' and x['properties'].get('geoid',np.nan) not in BadGeoDist]
```

```python
len(geoupdate)
```

```python
#Updating the geoid to be numeric
for i in geoupdate:
    i['properties']['geoid'] = int(i['properties']['geoid'])
```

```python
#Checking 1 record to verify
geoupdate[20]['properties']['geoid']
```

```python
geofile['features'] = geoupdate
```

```python
with open('dataset/CleanedData/cleaned-us-115th-congress.geojson','w') as outfile:
    outfile.write(json.dumps(geofile))
```

From the above climate opionion set, the following look like interesting variables:  
human - Estimated percentage who think that global warming is caused mostly by human activities  
humanOppose - Estimated percentage who think that global warming is caused mostly by natural changes in the environment
  
happening - Estimated percentage who think that global warming is happening  
happeningOppose - Estimated percentage who do not think that global warming is happening  

affectweather - Estimated percentage who somewhat/strongly agree that global warming is affecting the weather in the United States  
affectweatherOppose - Estimated percentage who somewhat/strongly disagree that global warming is affecting the weather in the United States  

worried - Estimated percentage who are somewhat/very worried about global warming  
worriedOppose - Estimated percentage who are not very/not at all worried about global warming 

congress - Estimated percentage who think Congress should be doing more/much more to address global warming  
congressOppose - Estimated percentage who think Congress should be doing less/much less to address global warming
