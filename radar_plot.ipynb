```
## merge : districtId(dis_event) && congress19(geoid)

new = pd.merge(dis_event, congress19,
                  left_on='districtId',
                  right_on='geoid', how= 'right')
```

```
new_red = new[['districtId','state','event','alerts_total','party','geoid']]
```

```
new_red = new_red.dropna()
```

```
new_red['districtId']= new_red['districtId'].astype(str)
```

```
new_red.info()
```

```
gkk = new_red.groupby(['event', 'party']).agg({'alerts_total': 'sum'})
```

```
gkk_event = new_red.groupby(['event']).agg({'alerts_total': 'sum'})
```
